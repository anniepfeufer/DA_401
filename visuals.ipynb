```python
import os
import os.path
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as pyplot
```

```python
total=pd.read_csv("data/total.csv")
grouped=total.groupby(["Tillage_st_x", "Year"]).agg({"Yield_bu_ac":"mean", "input_cost":"mean", "profit":"mean", "roi":"mean", "income_per_ac":"mean"})
```

```python
boxplot = total.boxplot(["profit"], by = ["Tillage_st_x"],
                     figsize = (16, 9),
                     showmeans = True,
                     notch = False)

boxplot.set_xlabel("Tilling Type")
boxplot.set_ylabel("Profit ($)")
boxplot.set_title("Profit Distribution by Tillage Type")
```

```python
boxplot = total.boxplot(["roi"], by = ["Tillage_st_x"],
                     figsize = (16, 9),
                     showmeans = True,
                     notch = False)

boxplot.set_xlabel("Tilling Type")
boxplot.set_ylabel("ROI ($)")
boxplot.set_title("ROI Distribution by Tillage Type")
```

```python
boxplot = total.boxplot(["Yield_bu_ac"], by = ["Tillage_st_x"],
                     figsize = (16, 9),
                     showmeans = True,
                     notch = False)

boxplot.set_xlabel("Tilling Type")
boxplot.set_ylabel("Yield (Bushels per Acre)")
boxplot.set_title("Crop Yield Distribution by Tillage Type")
```

```python
boxplot = total.boxplot(["input_cost"], by = ["Tillage_st_x"],
                     figsize = (16, 9),
                     showmeans = True,
                     notch = False )

boxplot.set_xlabel("Tilling Type")
boxplot.set_ylabel("Input Cost Required ($)")
boxplot.set_title("Cost Distribution by Tillage Type")
```

```python
grouped.reset_index(inplace=True)
no_till = grouped[grouped['Tillage_st_x'] == "No_till"].copy()
no_till.dropna(inplace=True)
alt = grouped[grouped['Tillage_st_x'] == "Alt"].copy()
alt.dropna(inplace=True)
chisel = grouped[grouped['Tillage_st_x'] == "Chisel"].copy()
chisel.dropna(inplace=True)
conv = grouped[grouped['Tillage_st_x'] == "Conv"].copy()
conv.dropna(inplace=True)
```

```python
pyplot.plot( no_till["Year"], no_till["Yield_bu_ac"], label="No Till")
pyplot.plot( alt["Year"], alt["Yield_bu_ac"], label="Alternate Till")
pyplot.plot( chisel["Year"], chisel["Yield_bu_ac"], label="Chisel Till")
pyplot.plot( conv["Year"], conv["Yield_bu_ac"], label="Conventional Till")
pyplot.xlabel("Year")
pyplot.ylabel("Average Bushels Per Acre")
pyplot.title("Crop Yield by Tilling Type over 45 years")
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot( no_till["Year"], no_till["input_cost"], label="No Till")
pyplot.plot( alt["Year"], alt["input_cost"], label="Alternate Till")
pyplot.plot( chisel["Year"], chisel["input_cost"], label="Chisel Till")
pyplot.plot( conv["Year"], conv["input_cost"], label="Conventional Till")
pyplot.xlabel("Year")
pyplot.ylabel("Input Cost ($)")
pyplot.title("Expenses by Tilling Type over 45 years")
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot( no_till["Year"], no_till["profit"], label="No Till")
pyplot.plot( alt["Year"], alt["profit"], label="Alternate Till")
pyplot.plot( chisel["Year"], chisel["profit"], label="Chisel Till")
pyplot.plot( conv["Year"], conv["profit"], label="Conventional Till")
pyplot.xlabel("Year")
pyplot.ylabel("Profit ($)")
pyplot.title("Profit Potential by Tilling Type over 45 years")
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot( no_till["Year"], no_till["roi"], label="No Till")
pyplot.plot( alt["Year"], alt["roi"], label="Alternate Till")
pyplot.plot( chisel["Year"], chisel["roi"], label="Chisel Till")
pyplot.plot( conv["Year"], conv["roi"], label="Conventional Till")
pyplot.xlabel("Year")
pyplot.ylabel("Return on Investment ($)")
pyplot.title("ROI by Tilling Type over 45 years")
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot( no_till["Year"], no_till["profit"], label="profit")
pyplot.plot( no_till["Year"], no_till["input_cost"], label="expenses")
pyplot.plot( no_till["Year"], no_till["income_per_ac"], label="income")
pyplot.xlabel("Year")
pyplot.ylabel("value $")
pyplot.title("no till financial assesment")
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot( conv["Year"], conv["profit"], label="profit")
pyplot.plot( conv["Year"], conv["input_cost"], label="expenses")
pyplot.plot( conv["Year"], conv["income_per_ac"], label="income")
pyplot.xlabel("Year")
pyplot.ylabel("value $")
pyplot.title("conv till financial assesment")
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot( alt["Year"], alt["profit"], label="profit")
pyplot.plot( alt["Year"], alt["input_cost"], label="expenses")
pyplot.plot( alt["Year"], alt["income_per_ac"], label="income")
pyplot.xlabel("Year")
pyplot.ylabel("value $")
pyplot.title("alt till financial assesment")
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot( no_till["Year"], no_till["income_per_ac"], label="no till-income")
pyplot.plot( conv["Year"], conv["income_per_ac"], label="conventional-income")
pyplot.plot( no_till["Year"], no_till["input_cost"], label="no till-expenses")
pyplot.plot( conv["Year"], conv["input_cost"], label="conventional-expenses")
pyplot.xlabel("Year")
pyplot.ylabel("value $")
pyplot.title("No Till vs Conventional Till Income and Expenses")
pyplot.legend()
pyplot.show()
```
